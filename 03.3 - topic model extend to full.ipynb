# Cargar topic model

```python
from bertopic import BERTopic

topic_model = BERTopic.load("covid_twitter_topic_model_2020-04-16")
```

```python
topic_model.get_topic_info().head(100)
```

# Aplicar topic model al conjunto completo de tuits

```python
topics_full = topic_model.transform(docs)
```
